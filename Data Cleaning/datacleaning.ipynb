```python
import numpy as np

# Define a function to map values to either country names or territory names, or keep them as NaN
def mapping(value):
    if pd.notna(value):
        country_name = d3_compatible_iso_country_codes.get(value)
        territory_name = d3_compatible_territory_mapping.get(value)
        
        if country_name:
            return country_name
        elif territory_name:
            return territory_name
        else:
            return np.nan
    else:
        return np.nan
    
df['Importer_d3c'] = df['Importer'].map(mapping)
df['Exporter_d3c'] = df['Exporter'].map(mapping)
df['Origin_d3c'] = df['Origin'].map(mapping)
```
